# 1.2 Creating Black/White Latino Per Tract Choropleth Map 

Below you'll find  Black/White Latino Per Tract Choropleth Map using hvplot with the "Melted_DF" data set from part 1.1 
    
This is a tool to demonstrate the spread of different Races/Ethnicities Across Philadelphia Including "Black and Latino/Hispanic" and "White and Latino/Hispanic"
    


</div>

```python
import altair as alt
import geopandas as gpd
import hvplot.pandas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import requests

import warnings

warnings.filterwarnings("ignore")



```

<div class="text-box">

```python


#| echo: false
#| output: true
#| code-fold: true

earth_tone_palette = [
"#ffffe5",
"#fff7bc",
"#fee391",
"#fec44f",
"#fe9929",
"#ec7014",
"#cc4c02",
"#993404",
"#662506"
]

import holoviews as hv
hv.extension('bokeh')

import geopandas as gpd
import hvplot.pandas
import holoviews as hv
hv.extension('bokeh')

melted_df = gpd.read_file("Melted_DF.geojson")

choropleth_map = melted_df.hvplot.polygons(
    'geometry', 
    color='Population',
    groupby='Race/Ethnicity',
    dynamic=False, 
    cmap= earth_tone_palette, 
    width=600,  
    height=600,  
    title='Choropleth of Black/White Latine Population Per Census Tract in Philadelphia',
    colorbar=True, 
    hover_cols=['tract','Median Household Income'] 
)

choropleth_map = choropleth_map.opts(
    frame_width=500, 
    show_frame=False,  
    xaxis=None,  
    yaxis=None   
)


choropleth_map

```

## Analysis 
The above choropleth Map can give a good general visualization of the spread of Latine communities across Philadelphia. We see that the census tracts closer to  NorthEast Philadelphia have a larger "White/Latine Population" than "Black/Latine Population". NorthEast Philadelphia is known to be an area with a large median socioeconomic status in comparison to the more Northern areas (CITE) where "Black and Latine/Hispanic" populations are located,  indicating a potential higher prevelance of "White/Latines" in areas with higher income. 

Part 1.3 will check the validity of the previous claim "potential higher prevelance of "White/Latines" in areas with higher income" through a comparison of the mean  household incomes in tracts with higher concentrations of Black/Latines than White Latines and vice-versa.

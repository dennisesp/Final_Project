# 2.1 Data Merging and Cleaning

In this part I,

1) Download the data from open data Philly, and the geojson demographic/tract dataframe from Part 1.1

2) Merge both of the data frames

3) Clean the data frame contaning columns that I want for the analysis

4) Export the cleaned dataframe for use in part 2.2 



```python
#| echo: true
#| code-fold: true

college_buildings=gpd.read_file("Universities_Colleges.geojson")

demographics_with_tracts= gpd.read_file("identity_with_tracts.geojson")

buildings_with_tract = gpd.sjoin(
    college_buildings,
    demographics_with_tracts.to_crs(college_buildings.crs),
    how='left',
    predicate="within",
)







columns_to_keep = ['NAME_left', 'ADDRESS', 'BUILDING_DESCRIPTION',
       'geometry', 'Black and Latino/Hispanic',
       'Total Population', 'Median Household Income', 
       'Bachelors Degree or Higher',  'Associates Degree',
       'Masters Degree', 'White and Latino/Hispanic',
       'tract']

buildings_with_tract_clean = buildings_with_tract[columns_to_keep]

buildings_with_tract_clean.to_file("buildings_per_tract.geojson")

buildings_with_tract_clean
```
